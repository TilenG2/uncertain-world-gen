```python
import utils
import ploting
```

```python
world_seed = 1234
data_seed = 123
number_of_features = 2
no_samples = 10**3

equal_classes = True
"""
equal_classes:
    if set to False the classes will have diffrent number of elements
    and the total number of elements is equal to no_samples

    if set to True the classes will have the same number of elements
    and the total number of elements is equal to no_samples * class_number
"""
feature_bounds = (0, 1) # set min an max value for all features
class_number = 3

# stoping criteria for building trees
endArea = 0.4 # the lower the number more complex are world ideal range [0, 0.4]
endThreshold = .2 # what the minimum width of cut can be .2 = 20% of the bounds
max_depth = 4 # when to stop building the tree 4 works well

# observerd world parameters i sudgets you generate thees yourself use generate_world_pandas and then your own method for 'ov' and 'unc'
errRange = 0.1
corr = 1

#generate_world returns a 2d np.array like
# Class, True Value 1, True Value 2, Observed Value 1, Uncertainty 1, Observed Value 2, Uncertainty 2
world = utils.generate_world(world_seed, data_seed, number_of_features,
                   no_samples = no_samples,
                   equal_classes = equal_classes,
                   feature_bounds = feature_bounds,
                   endArea = endArea,
                   endThreshold = endThreshold,
                   max_depth = max_depth,
                   class_number = class_number,
                   errRange = errRange,
                   corr = corr)
ploting.draw_combined_world(world)
```
